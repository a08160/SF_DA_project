```python
import pandas as pd

rent_data = pd.read_csv("C:\\Users\\TG\\코딩온 KDT\\시각화 프로젝트\\지역별_평균_월세.csv")
management_fee_data = pd.read_csv("C:\\Users\\TG\\코딩온 KDT\\시각화 프로젝트\\자치구별_관리비_평균.csv")

# 관리비 데이터에서 '서울시' 행 제거
management_fee_data = management_fee_data[management_fee_data['지역'] != '서울시']

# 컬럼명을 자치구명으로 통일
management_fee_data.rename(columns={'지역': '자치구명', '공용관리비용': '관리비(원)'}, inplace=True)

# 관리비 데이터를 만원 단위로 변환
management_fee_data['관리비(만원)'] = management_fee_data['관리비(원)'] / 10000

# 구를 기준으로 관리비/월세 결합
merged_data = pd.merge(rent_data, management_fee_data[['자치구명', '관리비(만원)']], on='자치구명', how='inner')

print(merged_data)

# 데이터 csv파일로 저장
merged_data.to_csv("구별_월세_관리비_평균.csv", index=False)
```

```python
import pandas as pd

rent_management_file = "C:\\Users\\TG\\코딩온 KDT\\시각화 프로젝트\\구별_월세_관리비_평균.csv"
income_spending_file = "C:\\Users\\TG\\코딩온 KDT\\시각화 프로젝트\\소득소비_전처리.csv"

rent_management_data = pd.read_csv(rent_management_file, encoding="utf-8-sig")
income_spending_data = pd.read_csv(income_spending_file, encoding="utf-8-sig")

# 3️데이터 병합 (구 기준으로 결합)
merged_data = pd.merge(income_spending_data, rent_management_data, left_on="구", right_on="자치구명", how="left")

# 주거 부담 비율 계산 (월세 + 관리비) / 소득
merged_data["총 주거비(만원)"] = merged_data["월세(만원)"] + merged_data["관리비(만원)"]
merged_data["주거 부담 비율"] = (merged_data["총 주거비(만원)"] / (merged_data["소득"] / 10000)).round(2)  # 소수점 2자리 반올림
```

```python
# 거주 적합도 점수 계산 + 소득만 반영, 소비는 제외
def calculate_score(row):
    base_score = 100  # 기본 점수
    income_factor = row["소득"] / merged_data["소득"].max()  # 소득 가중치
    housing_factor = 1 - row["주거 부담 비율"]  # 주거 부담이 낮을수록 유리
    
    score = base_score * income_factor * housing_factor  # 소비 반영 X
    return round(max(score, 0), 2)  # 최소 0점 유지, 소수점 2자리 반올림

merged_data["거주 적합도 점수"] = merged_data.apply(calculate_score, axis=1)

# 데이터 정리
final_columns = ["구", "동", "월세(만원)", "관리비(만원)", "소득", "주거 부담 비율", "거주 적합도 점수"]
final_data = merged_data[final_columns].sort_values(by="거주 적합도 점수", ascending=False)

# 데이터 저장
final_data.to_csv("최종_거주_적합도1.csv", index=False, encoding = "utf-8-sig")
```

```python
# '구'와 '동'을 유지하면서 숫자형 열의 평균을 계산
final_grouped = final_data.groupby(["구", "동"]).mean(numeric_only=True).reset_index()

print(final_grouped)
final_grouped.to_csv("찐막최종_거주_적합도.csv", index=False, encoding = "utf-8-sig")
```
